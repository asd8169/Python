# 데이터 다운로드 하기 - 1
: 인터넷에서 지정된 파일을 바로 내 PC에 저장하기

```python
import urllib.request
# 파이썬에서 인터넷 데이터를 받아 오는 기능들이 들어있는 라이브러리

# URL과 저장경로 지정하기.
url = "http://uta.pw/shodou/img/28/214.png"
savename = "../Data/test.png"

# 다운로드 하기

#request해서 가져오기 savename에 저장
# urlretrieve(이미지주소, 저장경로)

urllib.request.urlretrieve(url, savename)
print("저장 되었습니다!")
```

# 데이터 다운로드 하기 -- 02
: 인터넷에서 지정된 파일을 메모리에 올리고 저장하기


```python
import urllib.request

# URL과 저장경로 지정하기.
url = "http://uta.pw/shodou/img/28/214.png"
savename = "../Data/test2.png"


# 다운로드 하기

# 파일을 가져와서 수정할떄
mem = urllib.request.urlopen(url).read()

# 파일로 저장하기 

# 바이너리로 저장 wb
with open(savename, "wb") as f:
    f.write(mem)
    
print("저장 되었습니다!")
```

# 매개변수를 추가해 요청을 전송하는 방법 

기상청 RSS : http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp
        
URL 뒤에 매개변수를 입력하여 지역별로 일기 정보를 가져올 수 있음.

전국 : 108, 서울/경기도 : 109, 강원도 : 105, 제주도 : 184

ex) 서울/경기도

http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stdId=109

```python
import urllib.request
#파싱 임포트
import urllib.parse

API = "http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매개변수를 URL Encoding

values = {'stnId':'109'}
params = urllib.parse.urlencode(values)

# 요청 전용 URL 생성
url = API + "?" + params
print("url =", url)

data = urllib.request.urlopen(url).read()

# print(data)
#print(data) # 글자를 확인할 수 없음
text = data.decode('utf-8')
print(text)
```

# BeautifulSoup로 Scraping 하기
- BeautifulSoup : 간단하게 HTML과 XML에서 정보 추출
- Beautiful 설치 : 터미널에 pip install beautifulsoup4 입력

```python
# 기본 사용법 


from bs4 import BeautifulSoup

# HTML Sample 만들기
html = """
       <html>
           <body>
               <h1> Header </h1>
               <p> Line 1을 서술 </p>
           </body>           
        </html>
"""

print(html)
print("---------------")
# HTML 분석
# parser 방식으로 분리해라
soup = BeautifulSoup(html, 'html.parser')
print(soup)

# 원하는 부분 추출하기
# 단계별 추출
# 필요한 부분을 태그 이름으로 구분 html -> body -> h1으로 가져와라

h1 = soup.html.body.h1
p = soup.html.body.p

print("-----태그 가져오기-----")
print("")
print(h1)
print(p)


# text만 출력
# h1태그안 문자열 가져오기
print("")
print("-----문자열 가져오기-----")
print("h1 =", h1.string)
print("p =", p.string)
```

```python
# id로 요소를 추출하기

from bs4 import BeautifulSoup

# HTML Sample 만들기
html = """
       <html>
           <body>
               <h1 id = "title"> Header </h1>
               <p id = "body"> Line 1을 서술 </p>
           </body>           
        </html>
"""

print(html)
print("---------------")
# HTML 분석
# parser 방식으로 분리해라
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
title = soup.find(id='title')
body = soup.find(id='body')

print("")
print(title)
print(body)

# text만 출력
print("\n-----문자열 가져오기-----\n")
print('title =', title.string)
print('body =', body.string)

print('title =', title.text)
print('body =', body.text)
print('')
```

```python
# 여러개의 요수 추출하기
from bs4 import BeautifulSoup

# HTML Sample 만들기
html = """
       <html>
           <body>
               <ul>
                   <li><a href="http://www.naver.com">naver</a></li>
                   <li><a href="http://www.daum.net">daum</a></li>
               <h1 id = "title"> Header </h1>
               <p id = "body"> Line 1을 서술 </p>
           </body>           
        </html>
"""

print(html)


# HTML 분석
# parser 방식으로 분리해라
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
# 주소에 가장 근접해있는 a태그 찾기(a 태그의 경우 태그안 내용 같이 옴)
# find는 하나만 찾아주는애 all은 전부

links = soup.find_all('a')

#리스트에 담겨서 옴
print("\n----- 원하는 부분 추출하기-----\n")
print(links)

for link in links:
    href = link.attrs['href']
    text = link.string
    print(text, "==>, ",href)
    
```

```python
# 기상청 자료 활용하기
from bs4 import BeautifulSoup
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

# data가져오기
res = req.urlopen(url)

# BeautifulSoup으로 분석

soup = BeautifulSoup(res, 'html.parser')
# print(soup)

# 원하는 데이터 추출하기

title = soup.find('title').string
wf = soup.find('wf').string
# wf = wf.replace('<br />','')

print("\n------------------\n")

print(title)
print(wf)
for char in wf.replace('<br /',""):
    if char == '>':
        print("")
        continue
    print(char, end="")


# 데이터 정리하기

# 데이터 정리하기
list_wfs = list(wf)
except_char = ['<','b','r','/','>']
result = ''

for lwf in list_wfs:
    if lwf == '○':
        result +="\n"+lwf
    elif lwf not in except_char:
        result += lwf
 
        
print(title)
print("="*50)
print(result)
# type(wf)
# list(wf)
```

# css 선택자 사용하기
- soup.select_one() : css 선택자로 요소 하나를 추출
- soup.select() : css 선택자로 여러개를 리스트로 추출

```python
from bs4 import BeautifulSoup 

#HTML Sample
html = """
        <html>
            <body>
               <div id="meigen">
                    <h1>위키북스 도서</h1>
                  <ul class="items">
                    <li>유니티 게임 이펙트 입문</li>
                    <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
                    <li>모던 웹사이트 디자인의 정석</li>
                  </ul>
                </div>
            </body>
        </html>
"""

#HTML 분석하기
soup = BeautifulSoup(html,'html.parser')
      
#필요한 부분을 CSS 쿼리로 추출하기
#타이틀 부분 추출하기
h1 = soup.select_one("div#meigen > h1").string # id:#, class:. , tag : tag명 , 자손:> , 후손: ' '
print("h1=",h1)
      
#목록부분 추출하기
li_list=soup.select("div#meigen > ul.items > li")
for li in li_list:
    print("li=",li.string)
```

# csv파일 만들기

```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
```

```python
url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)
# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")
```

# 테스트

```python
#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
soup.select("#top_movies_main > div > table > tr > td > a")[0].string
```

# 적용

```python
li_lists = soup.select("#top_movies_main > div > table > tr > td > a")

count = 0

titleLists = [] # 영화제목 Lists
countLists = [] # Count Lists

for li in li_lists:
    count += 1
    print(count, li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countLists만으로 dataframe 생성
df = pd.DataFrame(data=countLists, columns=['index'])
df
```

```python
# 기존 dataframe에 titleLists 추가

df['Titles'] = pd.Series(titleLists)
df
```

```python
# csv로 저장
df.to_csv("../Data/rottenTomatoesList_2019.csv", encoding='utf-8', index=False)
```
